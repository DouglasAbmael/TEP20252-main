```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
turma_1 = np.random.normal(75, 8, 30)
turma_2 = np.random.normal(68, 12, 30)
turma_3 = np.random.normal(82, 5, 30)

dados = [turma_2, turma_2, turma_3]
```

```python
plt.figure(figsize=(8, 6))

dic = {
    'color': 'red',
    'linewidth': 1,
    'linestyle': '-'
}

plt.boxplot(dados, labels=['Turma 1', 'Turma 2', 'Turma 3'], patch_artist=True, medianprops=dic)
plt.title('Comparação da distribuiçao de notas entre turmas')
plt.ylabel('Notas')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
