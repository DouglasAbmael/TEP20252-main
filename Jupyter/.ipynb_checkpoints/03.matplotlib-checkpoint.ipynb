### 1) GRÁFICO DE LINHAS: Evolução da temperatura ao longo de uma semana
---------------------------------

```python
import matplotlib.pyplot as plt
```

```python
dias = ['seg', 'ter', 'qua', 'qui', 'sex', 'sab', 'dom']
temperaturas = [22, 20, 25, 23, 20, 27, 30]
```

```python
plt.figure(figsize=(8,4))
plt.plot(dias, temperaturas, marker='o', linestyle='--', color='r')
plt.title('Evolução da Temperatura Durante a Semana')
plt.xlabel('Dia da Semana') 
plt.ylabel('Temperatura (°C)')
plt.grid(True)

plt.show()
```

```python
# Eixo X: Meses que antecedem a eleição
meses = ['Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro']

# Eixo Y: Intenção de votos (em %) para cada candidato
# Os valores são fictícios, criados para mostrar tendências diferentes.
intencoes_collor = [5, 14, 25, 20, 19, 22]
intencoes_itamar = [10, 15, 21, 16, 14, 19]
intencoes_fhc    = [20, 22, 25, 28, 32, 36]

# --- Criação do Gráfico ---

# Definindo o tamanho da figura para uma melhor visualização
plt.figure(figsize=(10, 6))

# Linha para Fernando Collor
plt.plot(meses, intencoes_collor, marker='s', linestyle='--', color='green', label='Fernando Collor')

# Linha para Itamar Franco
plt.plot(meses, intencoes_itamar, marker='o', linestyle='-', color='blue', label='Itamar Franco')

# Linha para Fernando Henrique Cardoso
plt.plot(meses, intencoes_fhc, marker='^', linestyle='-.', color='red', label='FHC')


plt.title('Cenário Hipotético: Intenção de Votos para Presidente (1994)')

plt.xlabel('Mês da Pesquisa')
plt.ylabel('Intenção de Votos (%)')

plt.ylim(0, 50)
plt.legend(loc="upper right")
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

### 2) GRAFICO DE BARRAS: Comparar a quantidade de produtos vendidos em diferentes categorias
---------------

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
vendas = [120, 95, 210, 80]
```

```python
plt.figure(figsize=(8, 5))

plt.bar(categorias, vendas, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Vendas por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.ylim(0, 240)

plt.show()
```

### 3) HISTOGRAMA: analisar a distribuição das notas de uma turma de 100 alunos
---------------

```python
import numpy as np
np.random.seed(42)
notas = np.random.normal(70, 10, 100)

plt.figure(figsize=(8, 5))
plt.hist(notas, bins=10, edgecolor='black')
plt.title('Distribuição das Notas da Turma') 
plt.xlabel('Nota') 
plt.ylabel('Frequência (Nº de Alunos)')
plt.axvline(notas.mean(), color='red', linestyle='dashed', linewidth=2)
plt.legend(['Média das Notas'])

plt.show()
```

### 4) GRÁFICO DE PIZZA: visualizar a composição de uma carteira de investimentos
---------------

```python
ativos = ['Ações', 'Renda Fixa', 'Fundos Imobiliários', 'Ouro']
valores = [45, 30, 15, 10]

plt.figure(figsize=(7, 7))

explode = (0.1, 0, 0, 0)
plt.pie(valores, labels=ativos, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('Composição da Carteira de Investimentos')
plt.axis('equal')

plt.show()
```

### 5) GRÁFICO DE DISPERSÃO: analisar a relação entre a altura e o peso de um grupo de pessoas
---------------

```python
np.random.seed(0)
alturas = np.random.normal(170, 10, 50) 
pesos = (alturas * 0.5) + np.random.normal(0, 5, 50)

plt.figure(figsize=(8, 5))

plt.scatter(alturas, pesos, alpha=0.7, color='red', s=20)
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

plt.show()
```

### 6) GRÁFICO BOXPLOT: 
---------------

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
turma_a = np.random.normal(75, 8, 30)
turma_b = np.random.normal(68, 12, 30)
turma_c = np.random.normal(82, 5, 30)
dados_turmas = [turma_a, turma_b, turma_c]

plt.figure(figsize=(8, 6))

#d = {'color': 'g', 'linewidth': 1, 'linestyle': '-'}

plt.boxplot(dados_turmas, tick_labels=['Turma A', 'Turma B', 'Turma C'], patch_artist=True, medianprops=dict(color='g', linewidth=1, linestyle='-'))

plt.title('Comparação da Distribuição de Notas entre Turmas')
plt.ylabel('Nota')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
